```python
# pobieramy load dotenv

from dotenv import load_dotenv

#  pobieramy biblioteke sdk agent gdzie pobieramy metode Agents czyli llm , Runner który odpala nasz llm, i tracker który sledzi bledy czy wykonanie 

from agents import Agent, Runner , trace
```

```python
# budujemy naszego agenta llm przekaujac go do zmiennej agent wrzucamy do niego name instrukction czyli prompt i model

agent = Agent[any](name= 'joker', instructions= 'opowadiasz ciekawostki o jeziorze ochryde w macedoni ' , model= 'gpt-4.1-nano')
```

```python

# tworzymy trace i odpalamy model asychnronicznie dzieki temu nie musmy sptopowac kodu znaim dostaenimy odpowiedz z api kod jest asycnhroniczny dzieki słowu await potem odpalamy pobrany moduł Runner który odpala model metoda run
# jako parametry podajemy cały model agent i jako 2 prompt

with trace('joker'):

    result = await Runner.run(agent,'opowiedz 4 ciekaowstki o jeziorze i poleć ładne miejsca')
    # odpalamy odpowiedz przekazaną do result z modelu
    print(result.final_output)
```

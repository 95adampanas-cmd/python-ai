```python
# import 

from dotenv import load_dotenv

# teraz robimy import pod sdk 
# wpierw importujemy agents czyli llm z instrukcjami i toolsami zamiast budowac system prommpt + tools + petle while - tworzymy obiekt  Agent
# Runner - klasa która uruchamia agenta odpowiednik funkncji chat przyjmuje agenta i wiadomosc i history obsluguhe tool_calls - zwracaodpowiedz
# trace - narzedzie do debugowania i moniotorowania  weryfikuje jakie narzedzia uzyte ile to trfawło itp 

from agents import Agent, Runner , trace 
```

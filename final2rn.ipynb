```python
from dotenv import load_dotenv

load_dotenv(override=True)
import json
import os

openai_api_key = os.getenv('OPENAI_API_KEY')

print(openai_api_key)

from openai import OpenAI

myOpenAI = OpenAI() 
#

competitors = []
answers = []
```

```python
# lista zapytania, tworzymy zapytanie

question = 'utwórz pytanie dla sztucznej inteligencji, jakie są najlepsze dania na świecie, nie odpowiadaj, zadaj pytanie na które mają odpowiedzeć modele'

message = [{'role':'user','content': question}]
```

```python
# model dla utworzenia zapytania

response = myOpenAI.chat.completions.create(

    model = 'gpt-4',
    messages=message
)

finalquestion = response.choices[0].message.content

print(finalquestion)
```

```python
# zaputanie dla modeli

message = [{'role':'user', 'content': finalquestion}]
```

```python
#1 model 

response = myOpenAI.chat.completions.create(

    model = 'gpt-4.1-mini',
    messages=message
)

competitor = 'gpt-4.1-mini'
answer = response.choices[0].message.content

competitors.append(competitor)
answers.append(answer)

print(answer)
```

```python
#2 model

response = myOpenAI.chat.completions.create(

    model = 'gpt-4.1-nano',
    messages=message
)

competitor = 'gpt-4.1-nano'
answer = response.choices[0].message.content

competitors.append(competitor)
answers.append(answer)

print(answer)
```

```python
# grupoujemy po indexach w pętli odpowiedzi razem z modelami 
#Ten kod iteruje po dwóch listach naraz i od razu wypisuje parę: model → jego odpowiedź.
#Nic nie zapisuje ani nie buduje – służy tylko do natychmiastowego podglądu danych.
#➡️ Używasz tego, gdy chcesz tylko zobaczyć wyniki w konsoli i nie planujesz ich dalej przetwarzać. czyłi łączymy razem model + odpowiedz z tym samym idexem ale nachwie

for competitor,answer in zip(competitors,answers):
    print(competitor,answer)
```

```python

#Ten kod skleja wszystkie odpowiedzi w jeden duży tekst z numerami modeli.
#fektem jest jeden string, który można wysłać dalej (np. do judge-agenta) lub zapisać.
#➡️ Używasz tego, gdy chcesz przekazać WSZYSTKIE odpowiedzi jako jedną całość do kolejnego kroku.
togheter = ''

for index,answer in enumerate(answers):
    togheter += f'model nr {index +1}\n'
    togheter += answer 


print(togheter)
```

```python
judge_message = f"""oceniasz konkurs miedzy  {len(competitors)} konkuertami oceń który model poradził sobie lepiej, daj odpowiedź w json,

pytanie na które miały odpowiedzieć modele to {finalquestion}, oto odpoweidzi modeli {togheter} , uzasadnij wybór pokaż potem każego uczestnika"""

print(judge_message)
```

```python
# tworzymy model odpowiedzi

message = [{'role':'user', 'content': judge_message}]

response = myOpenAI.chat.completions.create(

    model = 'gpt-4',
    messages=message
)


finalAnswer = response.choices[0].message.content

print(finalAnswer)
```

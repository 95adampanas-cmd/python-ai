```python
# from to słowo kluczowe, keyword metoda uzywana w pythonie która mówi wskazuje z jakiego modułu będzie miał coś pobrać iternpeter
#from mówi w skrócie idz do konkretnego modułu
# dotenv to nazwa modułu, biblioteki, tutaj uzywamy go dow wczytywania pliku .env w którym przechowujemy klucz api do openAI 
#import to kolejne słowo kluczowe metoda która okresla co dokładnie mamy pobrac z biblioteki wskazanej przez from
#from wskazuje cała biblioteke import konkretny element biblioteki
#load_dotenv to gotowa funkcja, juz napisana, z biblioteki python-dotenv jej zadaniem jest odczytać plik .env  z biblioteki dotenv

from dotenv import load_dotenv

#tym kodem uruchamiamy funkcje load_dotenv z biblioteki dotenv, pobieramy zawartośc pliku .env który zawiera nasz klucz api, wczytujemy plik .env 
# i jego zawartosć
# override=True to argument dla funkcji, przekazuje jej jakies dane czy metode do uzycia w wyniku działania funkcju
#override= True to kod który zapewnia nam że dane z pliku .env zawsze zostaną nadpisane przez te z Api nawej jesli w ciele jakiejś funkcji
#isnieje inny klucz api to zostanie wykorzystany tez z pliku .env


load_dotenv(override=True)
```

```python
#import słowo klucz które pobiera daną rzecz z modułu lub moduł, tutaj bez from bo nie pobieramy 1 rzeczy z biblioteki tylko całą biblioteke
#os to moduł ze standardowej biblioteki pythona, kod, interpeter może dzięki niemu pobierać pliki czy dane 
# open_api_key to nazwa zmiennej 
#os.getenv getenv to funkcja w module os, która pobiera i zwraca do zmienenj w tym wypadku open_api_key, zawartość wskazanej zmiennej z pliku .env
# w naszym pliku mamy zmienną OPENAI_API_KEY do której przypisany jest klucz api, dzięki funkcji getenv pobieramy jej zawartosc i przekzujemy
# do zmiennej openai_api_key 
# teraz dzieki modułowi os który czyta pliki, metodą getenv czy tez fubkcja w naszej zmiennej mamy klucz api

import os 

openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set - please head to the troubleshooting guide in the setup folder")

    
    
```

```python
#wskazuje z jakiej biblioteki pobrzeiemy moduł
# openai to biblioteka pythona z openai do działania w ai z pythonem zawiera klasy itp pozwala na komunikacje z openai i modelami jak gpt5 itp
# import pobiera cos z biblioteki openai
#OpenAI to klasa z biblioteki openai, która pozwala np na utworzenie clienta, przez którego dzieki api kominokujemy sie i rozmwiamć i wymieniać dane
# w kodzie python  z modelem llm
from openai import OpenAI

# zaimportowaną klase OpenAI z modułu openai uruchamiamy i przeakzujemy do naszej zmiennej openaiVar

openaiVar = OpenAI()
```

```python
#messages to nazwa zmiennej, do któej przypisujemy liste wiadomosci dla modelu llm bo llm pracuje na podstawie histori wiadomosci
# role, klucz , słowo kluczowe które mówi llm-owi kto jest autorem wiadomosci są 3 opcje - user - uzytkownik - assistant - llm czyli jego odpowiedz
# i system - czyli instrukcje dla llm'a  
# content to słowo kluczowe które zawiera pytanie usera lub odpowiedz llm-a
messages = [{'role':'user','content': 'ile oczu mają koty'}] 
```

```python
# respone = tworzymy zmienną do ktorej przypiszemy odpowiedz llm'a na nasze zapytanie zawarte w masseges, odpowiedz zostana przechowana w response,
#mozna jej uzyc do kontuyjaacji rozmowy
#openaiVAr zawiera klasę, obiekt OpenAI uruchomioną i pobraną z modułu, biblioteki openai , obiekt ten słuzy do obsługi i wysyłania naszych zapytań
# do modelu llm 
# chat to moduł z klasy OpenAI który odpowiada za rozmowę tekstową 
#completions do pod moduł z klasy OpenAI który odpowiada za oobsługę wiadomosci testkowych 
# create to metodua, funkcja obiektu OpenAI, która wysyła wiadomosc do llma przez api, uruchania model llm, odbiera ją przetwarza i zwraca do naszej zmiennej
#model to argument metody create, odpowiada za wskazanie modelu z api openai, który zostamie uzyty 
#messages to kolejmy argument który oczekuje podania listy wiadomosci z role i content, które model przczyta i odpowie
response = openaiVar.chat.completions.create(

model="gpt-4.1-nano",
messages=messages

)

# response to nasza zmienne do której przekazujemy cały obiekt który zawiera model, odpowiedz 
# metoda choices to lista wszystkich odpowiedzi wygenerowanych przez model, moze byc ich kilka kazda ma swój index, zazwyczaj beirfzemy 1
#odpowiedz z indexem 0
#message to słowo klucz kóre zawiera odpowiedz z choises z indexem 0
#content to konkretna odpowiedz z llm'a
print(response.choices[0].message.content)
```

```python
 # tworzymy teraz długą rozmowę z kontekstem gdzie chat ma dostęp do poprzednich pytań 
# zaczynamy od pytania   rola to user, content wiadomosc tekstowa

messages = [{'role':'user','content': 'Wybierz jedną branżę, która może być interesującym polem do wdrożenia rozwiązań z użyciem Agentic AI' }]

# budujumemy model odpowiedzi

response = openaiVar.chat.completions.create(

    model= 'gpt-4o-mini',
    messages=messages

)
# odpowiedz przekazujemy do zmiennej i jatam rpzechowujemy

FirstAnswer = response.choices[0].message.content

print(FirstAnswer)

# teraz wazna rzecz do listy wiadomosci messages za pomocą metody appends dodajemy odpowiedz czyli zmieną FirstNaswer role wskazujemy assistant
# to 1 z 3 ról assistant oznacza że dodajemy odpowiedz, wiadomosc wyslaną przez llm. dzieki temu model roznioznia i rozumie czat, na wiadomosci 
# usera, polecenia system i odpowiedz czatu

messages.append({'role':'assistant','content':'FirstAnswer'})


# kolejne pytanie od usera

messages = [{'role':'user','content':'"Podaj konkretny problem w tej branży, który mógłby być rozwiązany przez Agentic AI."'}]

# odpowiedz na kolejne pytanie

response = openaiVar.chat.completions.create(

    model = 'gpt-4o-mini',
    messages=messages
)

SecAnswer = response.choices[0].message.content

print(SecAnswer)

# załączamy odpowiedź 

messages.append({'role':'assistant','content':SecAnswer})


# 3 pytanie 

messages = [{'role': 'user','content':'Zaproponuj rozwiązanie Agentic AI dla tego problemu.'}]

# 3 ODPOWIEDZ

response = openaiVar.chat.completions.create(

    model='gpt-4.1-nano',
    messages=messages
)

# zapis odpowiedzi

ThirdAnswer = response.choices[0].message.content

# jesli nie chcemy dalej konynować po prstu pirntujemy a jesli chcemy dalej dodajemy append jako assistant

print(ThirdAnswer)
```

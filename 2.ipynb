```python
# importujemy funkcje loaddotenv
from dotenv import load_dotenv 
# importujemy metode os
import os
# importujemuy klase openAI z api openai
from openai import OpenAI
# imporotujemy moduł Json
import json
# importujemy klase z api antropic twóców modelu llm claude


```

```python
load_dotenv(override=True)

openai_api_key = os.getenv('OPENAI_API_KEY')

myOpenAi = OpenAI()
```

```python

# TWORZYMY TERAZ PROMPT PODZIELONY NA 2 CZĘSCI 
# znak += oznacza dodaj tekst do stringa, i zapaimętaj istniejący



request = 'Proszę wymyśl trudne, złożone pytanie, które mógłbym zadać kilku modelom językowym (LLM), aby ocenić ich inteligencję'
request += 'Odpowiedz wyłącznie pytaniem, bez żadnych wyjaśnień'

# tworzymy liste messages

messages = [{'role': 'user', 'content': request}]

```

```python
# tworzymy zapytanie o request

response = myOpenAi.chat.completions.create(

    model='gpt-4.1-nano',
    messages=messages
)

question = response.choices[0].message.content

print(question)
```

```python
# tworzymy puste listy, tablice gdzie przekazemy wyniki, i nazwy modeli
competitors = []
answers = []
# tworzymy zapytanie, na które mają odpowiedzieć różne modele, zapytaniem jest odpowiedź bota na nasze pytanie , czyli rpzekazujemy zmienną z odpowiedzią bota
messages = [{'role': 'user', 'content': question}]
 
 
```

```python

# tworzymy odpowiedz modelu openai

modelname = 'gpt-4.1-nano'

# tworzymy model odpowiedzi na nasze question

response = myOpenAi.chat.completions.create(

    model='gpt-4.1-nano',
    messages=messages
)

answer = response.choices[0].message.content

# teraz dodajemy dzieki append modelname do arr competisions

competitors.append(modelname)
answers.append(answer)

```

```python
# 2 odpowiedż

modelname = 'gpt-4.1-mini'

response = myOpenAi.chat.completions.create(

    model = 'gpt-4.1-mini',
    messages=messages

)

answer = response.choices[0].message.content

competitors.append(modelname)
answers.append(answer)

```

```python
# printujemy tablice, gedzie przeakzalismy nazwy modeli i odpowiedzi

print(competitors)
print(answers)
```

```python

#  teraz uzywamy z bibilioteki typing modułu any który pozwala być danej zmiennej czymkolwiek

from typing import Any

# teraz pakujemy razem odpowiedzi do fomatu zip for to zapętlenie, competitor to nazwa zmiennej własnej do której trafią zogdnie z indesami
#elementy z tablicy competitors, answer to również nazwa zmiennej tam trafią zgodnie z indexami stringi z tablicy answers 
# dzięki in zip iterujemy po tablicach competiirors i tablicach answers i towirzymy pary z indexem 0,1,2 itp gdzie index 0 z competirors
# trafi do competitror[0] a index 0 z arr answers do answer[0] czyli dzięki for in zip tworzymy pętle do której trafią wyniki z tablic

for competitor, answer in zip(competitors,answers):
    print(f"Competitor: {competitor}/{answer}")
```

```python
# teraz grupujemy wyniki answers pętlą i metodą for

# tworzymy pusty string 
togheter = ''
# zapętlamy tabliuce answers, w pętli for index danego elementu przekazujemy do zmiennej index, elementy do answer dzieki emtodzie enumerate
for index, answer in enumerate(answers):
    togheter = f'#Response from model {index + 1}'
    togheter += answer

    print(togheter)
```

```python
# tworzymy stringa do oceny nasztch odpowiedzi modeli """" to string wielo lijnikowy
# len competitors wyprintuje wsztstkie modele w tbalict competitors
# question posiada wygenerowane zapytanie przez model gpt
# togheter zawiera odpowiedzi z answers

judge = f"""Oceniasz konkurs pomiędzy {len(competitors)} konkurentami.

Każdy model otrzymał następujące pytanie:

{question}

Twoim zadaniem jest ocena każdej odpowiedzi pod kątem jasności i siły argumentacji oraz uszeregowanie
ich od najlepszej do najgorszej.You MUST return ALL competitors in the ranking.


Odpowiedz w formacie JSON i tylko w formacie JSON, stosując następujący format:
{{"results": ["numer najlepszego uczestnika", "numer drugiego uczestnika", "numer trzeciego uczestnika", ...]}}

Oto odpowiedzi uczestników:

{togheter}

Teraz odpowiedz w formacie JSON, podając kolejność uczestników.
Nie dodawaj żadnych dodatkowych informacji, formatowania markdown ani bloków kodu."""
```

```python
print(judge)
```

```python
# tworzymy listę messages dla modelu żeby przekazać judges jako text na którym model ma pracowaćL przekazujemy zm

judgess_mess = [{"role":'user','content':judge}]
```

```python
# tworzymy model odpowiedzi

response = myOpenAi.chat.completions.create(

    model='gpt-4',
    messages=judgess_mess
)

results = response.choices[0].message.content

print(results)
```

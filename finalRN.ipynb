```python
# impor funkcji loaddotemv

from dotenv import load_dotenv

# uruchomienie

load_dotenv(override=True)

# import bilbioteki OS

import os

# przekazanie klucza api

openai_api_key = os.getenv('OPENAI_API_KEY')

print(openai_api_key)

#pobieramy klase openAI

from openai import OpenAI

myOpenAi = OpenAI()

import json
```

```python
# tworzymy listy do których wrzucimy elementy czyli nazwy modeli które konkurują i odpowiedzi

competitors = []
answers = []
```

```python
# tworzymy zapytanie na które odpowiadać będą modele ai również przez ai :O

request = 'utwórz pytanie na które będą musiały odpowiedzieć konkurujące ze sobą modele ai, dotyczące pozytywnych wpływów oraz negatywnych na społecześntwo,Odpowiedz wyłącznie pytaniem, bez żadnych wyjaśnień'

messages = [{'role':'user','content': request}]
```

```python
# tworzymy model odpowiedzi
model = 'gpt-4.1-mini'

response = myOpenAi.chat.completions.create(

 model = model,
 messages=messages
)

question = response.choices[0].message.content

print(question)
```

```python
# przekzujemy do messages wygnenrowane przez porzedni model zpaytanie na które teraz mają odpowiedzieć modele

```

```python
# 1 model 
messages = [{'role':'user','content': question}]
model = 'gpt-4o-mini'

response = myOpenAi.chat.completions.create(

    model = model,
    messages=messages
)

answer = response.choices[0].message.content

print(answer)

answers.append(answer)
competitors.append(model)

```

```python
# 2 model 
messages = [{'role':'user','content': question}]
model = 'gpt-4.1-mini'

response = myOpenAi.chat.completions.create(

    model= model,
    messages=messages
)

answer = response.choices[0].message.content

answers.append(answer)
competitors.append(model)

print(answer)
```

```python
# teraz tworzymy pętle po to by pogrupować razem elementy z tablic answers, competitros razem te które mają takie same indexy
# model, odpowiedz razem, potem kolejna para model, odpowiedz / do competitor trafi zmienna z indexem z arr competitros do naswer z answers
# uzywamy metody zip

for competitor,answer in zip(competitors,answers):
    print(competitor,answer)
```

```python
# teraz grupujemy odpowiedzi w pętli razem z indexami z tablicy answers 
togheter = ""
for index,answer in enumerate(answers):
    togheter = f'Result from model {index +1}'
    togheter += answer

    print(togheter)
```

```python

import os 
import json
from dotenv import load_dotenv
from openai import OpenAI

```

```python

#  uruchmamiamy dotenv

load_dotenv(override=True)

# pobieramy apo

openai_api_key = os.getenv('OPENAI_API_KEY')

# pobieramy klase openAI

myOpenAi = OpenAI()

```

```python

# tworzymy request dla openai, żeby wymyśliło pytanie na które mają potem odpoweidzieć modele

request =' Proszę wymyśl trudne, złożone pytanie, które mógłbym zadać kilku modelom językowym (LLM), aby ocenić ich inteligencję Odpowiedz wyłącznie pytaniem, bez żadnych wyjaśnień'

message = [{'role':'user', 'content': request}]
```

```python

# tworzymy model odpowiedzi

response = myOpenAi.chat.completions.create(

    model='gpt-4o-mini',
    messages=message
)

question = response.choices[0].message.content

print(question)
```

```python
# tworzymy arrays gdzie w 1 przechowamy nazwy modeli w 2 ich odpowiedzi

modelNames = []
answers = []
```

```python

# tworzymy zapytanie ze stworzonym zpaytaniem przerz czat w zmiennej question

messages = [{'role':'user','content':question}]

# tworzymy 1  odpowiedź

response = myOpenAi.chat.completions.create(

    model='gpt-4o-mini',
    messages=messages
)

answer = response.choices[0].message.content


modelName = 'gpt-4o-mini'

modelNames.append(modelName)
answers.append(answer)
```

```python
# tworzymy 2  odpowiedź

response = myOpenAi.chat.completions.create(

    model='gpt-4.1-mini',
    messages=messages
)

answer = response.choices[0].message.content


modelName = 'gpt-4.1-mini'

modelNames.append(modelName)
answers.append(answer)
```

```python
# pobieramy typing any

from typing import Any

# pętla for która zapętla arrays modelNames i answers , nastepnie in zip który w pętli po 2 tablicach pobiera z inch wartosci o tych samych idnexach
# przekazuje je do zmiennych model, answer
for model,answer in zip(modelNames,answers):
    print(f'competitor: {model}/n{answer}')

```

```python
print(modelNames)
print(answers)
```

```python
# tworzemy pustą zmienną togheter

togheter = ""

# tworzymy pętle for dodajemy zmienne index, do której trafi index danej odpowiedzi z tablicy answers a answer zawiera dany ciąg stringów z tablicy
# czyli index 0 + dana odpowiedz z tablicy bo wykonujemy pętle na tablicy answers metoda enumerate która rozbija tablice answers na stringi i wyciąga
# indexy, dzieje się to bo jestemsy w petli for

for index, answer in enumerate(answers):
    togheter += f"# Response from model {index+1}n\n"
    togheter += answer + "n\n"  
```

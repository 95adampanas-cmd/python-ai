```python
# importujemy biblioteke dotenv z któej to wyciągamy funkcje load_dotnev któr

from dotenv import load_dotenv

# uruchamiamy funkce load_dotenv, której zadaniem jest pobrać zawartosc pliku .env

load_dotenv(override=True)
```

```python

# importujemy biblioteke os kóra pozwala na odczytywanie plików, dla nas kluczowe bo pozwoli pobrac api openai

import os 

# do naszej zmiennej przekazujemy klucz api dzieki metodzie funkcji getenv z biiblioteki os, zwracamy zawartosc .env w którym znajduje się zmienna
# openai_api_key i jej zawratosc zwracamy do naszej czyli klucz api 
openai_api_key = os.getenv('OPENAI_API_KEY')


```

```python
# Z api open ai, importujemy klasę OpenAI, która pozwoli nam utworzyc klienta dzieki któremu będziemy w stanie komukiować kod z modelem llm

from openai import OpenAI


# pobieramy klase OpenAI uruchamiamy i przekazujemy do naszej zmiennej, dzieki temu będziemy w stanie porozumieć się z llm-em dzieki naszemu kodowi

openai = OpenAI()
```

```python
# tworzymy liste messages gdzie okreslamy role oraz tworzymy zapytanie content to moduł dla openai

messages = [{'role': 'user',"content": 'ile jest 2+2'}]


```

```python

# tworzymy klienta do przetwarzania zapytań i uruchomienia llm'a z wyborem silnika przekazujemy go do zmiennej,która będzie zawierała nasza odpowiedz
#z klasy open ai wynieramy modul odpowiedzialny za rozumienie tekstu pisanego
response = openai.chat.completions.create(

model='gpt-4.1-mini',
messages=messages

)

# printujemy nasz response wybieramy 1 z odpowiedzi zapisanych w chocies z indexem 0 z modułu message i neirzemy z niego content

print(response.choices[0].message.content)
```

```python
# tworzymy zapytanie - moduł rozmowy
messages = [{'role': 'user','content': 'Wybierz jedną branżę, która może być interesującym polem do wdrożenia rozwiązań z użyciem Agentic AI.'}]


response = openai.chat.completions.create(
    model = 'gpt-4.1-nano',
    messages=messages
)

fAnswer = response.choices[0].message.content

print(fAnswer)

messages.append({'role':'assistant','content': fAnswer})
messages.append({'role': 'user', 'content': 'Podaj konkretny problem w tej branży, który mógłby być rozwiązany przez Agentic AI.'})


response = openai.chat.completions.create(
    model = 'gpt-4.1-nano',
    messages=messages
)

sAnswer = response.choices[0].message.content

messages.append({'role': 'assistant', 'content': sAnswer})
messages.append({'role': 'user', 'content': 'Zaproponuj rozwiązanie Agentic AI dla tego problemu'})

print(sAnswer)

response = openai.chat.completions.create(
    model = 'gpt-4.1-nano',
    messages=messages
)


tAnswer = response.choices[0].message.content

print(tAnswer)


```

```python

# importujemy klucze api 1 to klucz usera 2 to klucz token
# przekazuejmy je do zmiennych za pomocą nazwy os.getenv i nazwy klucza api
pushover_user = os.getenv("PUSHOVER_USER")
pushover_token = os.getenv("PUSHOVER_TOKEN")

# dodajemy url na który bedziemy wysyłac kod

pushover_url = "https://api.pushover.net/1/messages.json"
```

```python

# piszemy funkcje ktora pozwoli nam wysłac wiadomosc na serwer jako parametr przekazujemy zmienna message która jest listą zmiennych i wartosci przypisanych do nich

def push(message):

    # printujemy wiadomość wysłana na serwer
    print(f'push : {message}')

    # do zmiennej przekazujemy wymagane przez api dane / user api, token api i message
    allData = {'user' : pushover_user , 'token' : pushover_token , "message" : message}

    # bibloteką request z metoda post wysyłamy wiadomośc na serwer, metoda ma 2 argumenty 1 to url serwera z pushover url 2 to lista danych allData

    requests.post(pushover_url, data = allData)
```

```python
# tworzymy funkcje/ narzędzie które zostanie uzyte, wywołane przez kolejne funkcje kodu, to narzędzie zbiera dane o uzytkowniku który wysyła nam wiadomość
# funkcja przyjmuje 3 argumenty email, name i notes / narzędzie zostanie uruchomione jesli nasz kod/ model / narzędzie bedzie w stanie odpowiedziec na pytanie usera
# wówczas uruchomi funkcje push z dodatkowymi informacjami

def record_user_details(email, name = 'not provided', notes = 'not provided'):

    # uruchamiamy funkcje push 
    push(f' record send from {email}, by {name} with notes {notes} ')

    # zwracamy info do modelu który wywoła funkcje, o tym ze record został dodany

    return {'recorded' : 'ok'}

```

```python


# kod z sdk

from agents import Agent, Runner, function_tool
from dotenv import load_dotenv
import csv

load_dotenv()

# zamiast pisać JSON dla każdego narzędzia - dekorator @function_tool
@function_tool
def save_email(company: str, email_content: str, subject: str) -> str:
    """Zapisuje wygenerowany mail do pliku CSV"""
    with open('mail.csv', 'a', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([company, email_content, subject])
    return 'saved'

@function_tool
def repeat_mail(company: str, reason: str) -> str:
    """Zapisuje odrzucony mail z powodem"""
    with open('repeatmail.csv', 'a', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([company, reason])
    return 'recorded'

# zamiast system_prompt jako string przekazujesz do Agent()
agent = Agent(
    name="Mail Generator",
    instructions="""Jesteś asystentem do generowania maili outreachowych dla Last Agency...""",
    tools=[save_email, repeat_mail],
    model="gpt-4.1-mini"
)

# zamiast funkcji chat() z pętlą while - Runner.run()
import asyncio

async def main():
    msg = """Firma: Tebim Pro
Branża: agencja e-commerce
Opis: tworzą sklepy internetowe na PrestaShop"""

    result = await Runner.run(agent, msg)
    print(result.final_output)

asyncio.run(main())
```
